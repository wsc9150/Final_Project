```python
# 단어 리스트 만들기
def make_list(df) :
    keyword_list = []

    for i in range(len(df)):
        temp = []
        
        for j in range(len(df['keyword'][i].split('\''))):
            if j % 2 == 1:
                temp.append(df['keyword'][i].split('\'')[j])
        keyword_list.append(temp)
    
    return keyword_list
```

```python
word_list=make_list(test)
```

```python
word_list
```

```python
from gensim.models.word2vec import Word2Vec
```

```python
embedding_model = Word2Vec(word_list, size=100, window = 2, min_count=2, workers=5, iter=1000, sg=1, sample=1e-3)
```

```python
from sklearn.cluster import KMeans

word_vectors = embedding_model.wv.syn0 # 어휘의 feature vector
# num_clusters = int(word_vectors.shape[0]/500) # 어휘 크기의 1/5나 평균 5단어
# print(num_clusters)
# num_clusters = int(num_clusters)
```

```python
word_vectors 
```

### 알파벳 8개

```python
kmeans_clustering = KMeans(n_clusters=8)
idx = kmeans_clustering.fit_predict(word_vectors)

idx = list(idx)
names = embedding_model.wv.index2word
word_centroid_map = {names[i]: idx[i] for i in range(len(names))}
```

```python
words = [[] for _ in range(8)]

for c in range(8):
    # 클러스터 번호를 출력
    print("\ncluster {}".format(c))
    
    cluster_values = list(word_centroid_map.values())
    for i in range(len(cluster_values)):
        if (cluster_values[i] == c):
            words[c].append(list(word_centroid_map.keys())[i])            
    # print(words)
```
