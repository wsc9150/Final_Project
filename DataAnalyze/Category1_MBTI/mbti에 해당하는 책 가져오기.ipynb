```python
# 책 데이터 불러오기

mbti_book=pd.read_excel('./data/final/mbti_book_revised.xlsx')
```

```python
# 단어 리스트 만들기
def make_list(df) :
    keyword_list = []

    for i in range(len(df)):
        temp = []
        
        for j in range(len(df['keyword'][i].split('\''))):
            if j % 2 == 1:
                temp.append(df['keyword'][i].split('\'')[j])
        keyword_list.append(temp)
    
    return keyword_list
```

```python
mbti_book['keyword'] = make_list(mbti_book)
```
