```python
mbti = ['ISTJ', 'ISFJ', 'INFJ', 'INTJ', 'ISTP', 'ISFP', 'INFP', 'INTP', 'ESTP', 'ESFP', 'ENFP', 'ENTP', 'ESTJ', 'ESFJ', 'ENFJ', 'ENTJ']
```

```python
# 군집의 키워드와 성격의 키워드 각각을 비교하며 있는지 없는지 검사
mbti_cluster_keyword = {}

for i in tqdm_notebook(range(len(cluster.columns))) :
    check = [0 for _ in range(16)]
    
    for j in range(len(cluster[i])) :
        
        for a in range(len(mbti_df.columns)) :
            for b in range(len(mbti_df[mbti[a]])) :
                
                # 결측치는 그냥 넘어가도록 설정
                if cluster[i][j] is np.nan or mbti_df[mbti[a]][b] is np.nan :
                    break
                    
                # print(cluster[i][j])
                # print(mbti_df[mbti[a]][b])
                
                # 군집 키워드에 성격 키워드가 포함되어 있는 경우 출력
                if mbti_df[mbti[a]][b] in cluster[i][j] :
                    print('겹치는 성격 유형 키워드 :', mbti_df[mbti[a]][b])
                    print('군집에 속한 단어 :', cluster[i][j])
                    print('성격 유형 :', mbti[a])
                    print('군집 :', i)
                    print()
                    check[a] += 1
                    
                    if mbti[a] not in mbti_cluster_keyword.keys() :
                        mbti_cluster_keyword[mbti[a]] = [cluster[i][j]]
                    else :
                        mbti_cluster_keyword[mbti[a]].append(cluster[i][j])
    
    print(f'군집 {i} 에 포함된 각 성격유형 키워드 수 : {check}')
    print()
```

## 도서키워드와 MBTI군집키워드 매칭

```python
# 단어 리스트 만들기
def make_list(df) :
    keyword_list = []

    for i in range(len(df)):
        temp = []
        
        for j in range(len(df['keyword'][i].split('\''))):
            if j % 2 == 1:
                temp.append(df['keyword'][i].split('\'')[j])
        keyword_list.append(temp)
    
    return keyword_list
```

```python
# 데이터프레임의 keyword 컬럼의 한 행씩 접근
for i in tqdm_notebook(range(len(mbti_book_revised_df))) :
    sum_list = []
    
    # mbti 성격 유형 아름(문자열) 순차적 접근
    for a in range(len(mbti)) :
        count_list = []
        
        # 한 행의 키워드 리스트에 하나의 키워드씩 접근
        for j in range(len(mbti_book_revised_df['keyword'][i])) :
            count = 0
            
            # 한 성격유형에 해당되는 키워드 한개씩 접근
            for b in range(len(mbti_cluster_keyword[mbti[a]])) :
                # 서로 속하는 여부 확인
                if (mbti_book_revised_df['keyword'][i][j] in mbti_cluster_keyword[mbti[a]][b]) or (mbti_cluster_keyword[mbti[a]][b] in mbti_book_revised_df['keyword'][i][j]) :
                    count += 1
    
            count_list.append(count)
        
        sum_list.append(sum(count_list))
    
    max_count = max(sum_list)
    print('sum_list =', sum_list)
    print('max_count =', max_count)
    
    for j in range(len(sum_list)) :
        if sum_list[j] == max_count :
            if j == 0 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                ISTJ_book_list_df = ISTJ_book_list_df.append(temp, ignore_index = True)
            if j == 1 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                ISFJ_book_list_df = ISFJ_book_list_df.append(temp, ignore_index = True)
            if j == 2 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                INFJ_book_list_df = INFJ_book_list_df.append(temp, ignore_index = True)
            if j == 3 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                INTJ_book_list_df = INTJ_book_list_df.append(temp, ignore_index = True)
            if j == 4 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                ISTP_book_list_df = ISTP_book_list_df.append(temp, ignore_index = True)
            if j == 5 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                ISFP_book_list_df = ISFP_book_list_df.append(temp, ignore_index = True)
            if j == 6 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                INFP_book_list_df = INFP_book_list_df.append(temp, ignore_index = True)
            if j == 7 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                INTP_book_list_df = INTP_book_list_df.append(temp, ignore_index = True)
            if j == 8 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                ESTP_book_list_df = ESTP_book_list_df.append(temp, ignore_index = True)
            if j == 9 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                ESFP_book_list_df = ESFP_book_list_df.append(temp, ignore_index = True)
            if j == 10 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                ENFP_book_list_df = ENFP_book_list_df.append(temp, ignore_index = True)
            if j == 11 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                ENTP_book_list_df = ENTP_book_list_df.append(temp, ignore_index = True)
            if j == 12 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                ESTJ_book_list_df = ESTJ_book_list_df.append(temp, ignore_index = True)
            if j == 13 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                ESFJ_book_list_df = ESFJ_book_list_df.append(temp, ignore_index = True)
            if j == 14 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                ENFJ_book_list_df = ENFJ_book_list_df.append(temp, ignore_index = True)
            if j == 15 :
                temp = mbti_book_revised_df.loc[i]
                temp02 = pd.Series(sum_list[j], index = ['keywordCnt'])
                temp = temp.append(temp02)
                ENTJ_book_list_df = ENTJ_book_list_df.append(temp, ignore_index = True)
```
