```python
# 단어 리스트 만들기
def make_list(df) :
    keyword_list = []

    for i in range(len(df)):
        temp = []
        
        for j in range(len(df['keyword'][i].split('\''))):
            if j % 2 == 1:
                temp.append(df['keyword'][i].split('\'')[j])
        keyword_list.append(temp)
    
    return keyword_list
```

```python
# 도서관 데이터와 서점 데이터를 합친 데이터
# 공통 컬럼만 뽑아내어 concat
# 중복 제거
new_final_df = pd.read_csv('new_final.csv')
```
